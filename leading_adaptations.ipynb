```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_dataset
```

```python


df = pd.read_csv('ai_adoption_dataset.csv')
print(df.head())
print(df.info())
```

```python
df
```

```python
df_leading_industries = df['industry'].value_counts().head(10).reset_index(name='Count')

df_leading_industries
```

```python
df_leading_regions = df['country'].value_counts().head(10).reset_index(name='Count')
df_leading_regions
```

```python
fig, ax = plt.subplots(2,1)

sns.barplot(data=df_leading_industries, ax=ax[0], hue='Count', x='Count',y='industry', palette='dark:b_r')
ax[0].set_title('Top 10 Leading Industries in AI Adaptation')
ax[0].set_xlabel('Count')
ax[0].set_ylabel('')
ax[0].get_legend().set_visible(False)
ax[0].set_xlim(0,18500)

sns.barplot(data=df_leading_regions, ax=ax[1], hue='Count', x='Count',y='country', palette='dark:b_r')
ax[1].set_title('Top 10 Leading Regions in AI Adaptation')
ax[1].set_xlabel('Count')
ax[1].set_ylabel('')
ax[1].get_legend().set_visible(False)
ax[1].set_xlim(0,15000)

fig.tight_layout()
plt.show()

```
